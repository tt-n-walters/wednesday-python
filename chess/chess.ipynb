```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

```python
df = pd.read_csv("games.csv")
```

```python
df.victory_status.value_counts()

wins_all = df.winner.value_counts()
white_win_ratio = wins_all.white / wins_all.black

wins_select = df[df.victory_status != "outoftime"].winner.value_counts()
white_win_ratio_select = wins_select.white / wins_select.black

figure, axes = plt.subplots(ncols=2, sharey=True)
wins_all.plot(kind="bar", ax=axes[0], figsize=(16, 4), title=white_win_ratio)
wins_select.plot(kind="bar", ax=axes[1], figsize=(16, 4), title=white_win_ratio_select)
```

```python
average_turns = round(df.turns.mean(), 1)
print(average_turns)
```

```python
white_ids = df.white_id.unique()
black_ids = df.black_id.unique()
ids = set(black_ids)
for id in white_ids:
    ids.add(id)
print(len(ids), "players")
```

```python
games_played = {}
for i, id in enumerate(ids):
    played_filter = df.white_id == id
    played = played_filter.sum()
    games_played[id] = played
    print(i, end="\r")
```

```python
print(games_played)
games_played_series = pd.Series(games_played)
print(games_played_series)
```
