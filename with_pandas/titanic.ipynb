```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-white")
```

```python
data = pd.read_csv("titanic_complete.csv")
```

## Passenger ages

```python
genders = data.gender.value_counts()
genders.plot(kind="pie")
```

## Passenger fares by gender

```python
males = data.gender == "male"
females = ~males

figure, axes = plt.subplots(ncols=2, sharey=True)
data[males].fare.plot(kind="hist", ax=axes[0], bins=40, ylim=(-10, 450), figsize=(16, 4))
data[females].fare.plot(kind="hist", ax=axes[1], bins=40, ylim=(-10, 450), figsize=(16, 4), style={"color":"r"})
```
